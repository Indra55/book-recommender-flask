```python
rating_with_books=ratings.merge(books,on="ISBN")
```

```python
rating_with_books.shape
```

```python
num_rating=rating_with_books.groupby('title')['rating'].count().reset_index()
```

```python
num_rating.rename(columns={"rating":"num_of_rating"},inplace=True)
```

```python
final_rating=rating_with_books.merge(num_rating,on='title')
```

```python
final_rating= final_rating[final_rating['num_of_rating']>=50]
```

```python
final_rating.sample(10)
```

```python
final_rating.drop_duplicates(['user_id','title'],inplace=True)
```

```python
final_rating.shape
```

```python
book_pivot = final_rating.pivot_table(columns='user_id',index='title',values='rating')
```

```python
book_pivot
```

```python
book_pivot.fillna(0,inplace=True)
```

```python
from scipy.sparse import csr_matrix
```

```python
book_sparse=csr_matrix(book_pivot)
```

```python
book_sparse
```

```python
from sklearn.neighbors import  NearestNeighbors
```

```python
model=NearestNeighbors(algorithm='brute')
```

```python
model.fit(book_sparse)
```

```python
distance, suggestion = model.kneighbors(book_pivot.iloc[254,:].values.reshape(1,-1),n_neighbors=6)
```
